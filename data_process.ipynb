```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
data = pd.read_csv("movies_metadata.csv")
data.head(5)
```

```python
# Load data into a DataFrame
df = pd.DataFrame(data)
# Function to convert the JSON-like string to a comma-separated string of genre names
def convert_genres(genre_str):
    try:
        # Parse the JSON-like string
        genre_list = json.loads(genre_str.replace("'", "\""))
        # Extract the genre names and join them with commas
        return ', '.join(genre['name'] for genre in genre_list)
    except (json.JSONDecodeError, TypeError):
        return None  # Return None if parsing fails

# Apply the function to the genres column
df['tag'] = df['genres'].apply(convert_genres)

# Display the DataFrame
out_df = df[['title', 'tag']]
print(df[['title', 'tag']].head())
```
